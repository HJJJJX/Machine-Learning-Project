<font color="blue"> Background:</font> Plugs will be set into the well when a well is completed. Then we will do the staged hydralic fracture to create the channel for oil and gas to move. After finishing the hydralic fracture, all the plugs will be drilled out. There is an increased uncertainty in bottom hole pressure experienced during drillout. And this uncertainty makes it diffcult to decide the drilling parameters including the average pump difference.

<font color="blue"> Goal:</font> In this work, we build a machine learning model to predict the average pump difference. We frist do the feature engineering for this dataset. Then test different models to find the optimal one that gives the best accurate predictions. With the best model, we tried multiple ways to optimize the parameters to reduce the overfit and improve the accuracy. Finally, we build a model which can best predict the results.

```python
num_fea = list(X.select_dtypes('number'))
cat_fea = list(X.select_dtypes('object'))

for column in num_fea:
    column_median = X[column].median()
    X[column].fillna(column_median,inplace=True)
    column_median_pred = X_pred[column].median()
    X_pred[column].fillna(column_median_pred,inplace=True)
```



*   <font color="blue"> Ranking the features based on the importance:</font>



```python
# Get the absolute values of principal component loadings
loadings_abs = abs(pca.components_)

# Create a DataFrame to display the importance of each feature
feature_importance = pd.DataFrame(loadings_abs.T, columns=[f'PC{i}' for i in range(1, 32)], index=num_fea + cat_fea)

# Sum the absolute loadings across principal components for each feature
feature_importance['Total Importance'] = feature_importance.sum(axis=1)

# Sort the DataFrame by total importance in descending order and select the top 8 features
top_features = feature_importance['Total Importance'].nlargest(31)

# Display the top 8 features with the order of importance
print(top_features)
```

<font color="blue"> KNN Regressor </font>



*   <font color="blue"> Check the optimum n_neighbor value with Cross Validation </font>


```python
# Configure the K-Fold cross-validation
kfold = KFold(n_splits=5, shuffle=True, random_state=2024)

# Range of `n_neighbors` values to evaluate
neighbors_range = range(1, 21)

# List to store average MSEs for each value of `n_neighbors`
average_mses = []

for n_neighbors in neighbors_range:
    knn = KNeighborsRegressor(n_neighbors=n_neighbors)

    # Perform cross-validation
    mse_scores = cross_val_score(knn, X, Y, cv=kfold, scoring='neg_mean_squared_error')

    # Convert scores to positive and calculate the average MSE
    mse_scores_positive = -mse_scores
    average_mse = np.mean(mse_scores_positive)
    average_mses.append(average_mse)

# Determine the best `n_neighbors` based on the lowest MSE
optimal_n_neighbors = neighbors_range[np.argmin(average_mses)]
print(f"Optimal number of neighbors: {optimal_n_neighbors}")
```

```python
plt.plot(neighbors_range, average_mses, marker='o', linestyle='-', color='blue')
plt.title('KNN Varying number of neighbors')
plt.xlabel('Number of Neighbors')
plt.ylabel('Average MSE')
plt.show()
```

```python
# Fit the train data with the optimal n_neighbor value
knn = KNeighborsRegressor(n_neighbors=optimal_n_neighbors)
knn.fit(X_train, y_train)
```
